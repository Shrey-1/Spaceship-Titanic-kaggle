```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
train.shape
train.info()
```

```python
train.head()
```

```python
train["PassengerId"]=train.PassengerId.apply(lambda x: x.split("_")[0])
train.drop(columns=['Cabin'],inplace=True)
```

```python
train.drop(axis=1,columns="Name",inplace=True)
train.PassengerId.astype(int)
```

```python
o=list()
for i in train.columns:
    if(train.dtypes[i]=="object"):
        o.append(i)
```

```python
train['PassengerId']=train.PassengerId.astype(int)
```

```python
la=LabelEncoder()
train['HomePlanet']=la.fit_transform(train['HomePlanet'])
train['CryoSleep']=la.fit_transform(train['CryoSleep'])
train['Destination']=la.fit_transform(train['Destination'])
train['VIP']=la.fit_transform(train['VIP'])
train['Transported']=la.fit_transform(train['Transported'])
```

```python
train.head()
```

```python
test.head()
```

```python
test["PassengerId"]=test.PassengerId.apply(lambda x: x.split("_")[0])
test.drop(columns=['Cabin'],inplace=True)
test.drop(axis=1,columns="Name",inplace=True)
test['PassengerId']=test.PassengerId.astype(int)
```

```python
test['HomePlanet']=la.fit_transform(test['HomePlanet'])
test['CryoSleep']=la.fit_transform(test['CryoSleep'])
test['Destination']=la.fit_transform(test['Destination'])
test['VIP']=la.fit_transform(test['VIP'])
```

```python
test.head()
```

```python
y=train["Transported"]
train.drop(columns="Transported",axis=1,inplace=True)
```

```python
train.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train, y, random_state=0)
```
